### ① 取出今天所有的發文

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.ettoday.net/news/news-list.htm'
resp = requests.get(url=URL)
```

```python
soup = BeautifulSoup(markup=resp.text, features='html.parser')
nodes = soup.select(selector = 'div h3')
for i in nodes:
    print(i.text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
content = []
for i in nodes:
    if '2021/05' in i.text:
        content.append(i.text)
```

```python
cato = []
for i in range(len(content)):
    cato.append(content[i][16:18])
```

```python
news = pd.DataFrame(zip(cato,content),columns=['分類', '標題'])
```

### ③ 哪一個類別的文章最多

```python
news['分類'].value_counts()
# 生活類
```
