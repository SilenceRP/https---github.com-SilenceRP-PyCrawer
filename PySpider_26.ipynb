```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.ettoday.net/news/news-list.htm'
resp = requests.get(url=URL)
```

```python
soup = BeautifulSoup(markup=resp.text, features='html.parser')
nodes = soup.select(selector = 'h3 a')
for i in nodes:
    print(i.text)
```
