## 練習從東森新聞雲網站中，爬取新聞摘要及詳細內容。
## 學習利用Selenium模擬人為操作，更新動態網頁後爬取新聞內容。

```python
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

### 以財經新聞為例，先連結到財經新聞網頁，複製其URL。

```python
URL = "https://www.ettoday.net/show_roll.php"  #財經新聞
formdata = {'offset':'2','tPage':'2','tFile':'104.json',
             'tOt':'0','tSi':'24','tAr':'0'}    
```

```python
title = []
summary = []

for i in range(1,6):
    formdata['offset'] = str(i)
    resp = requests.post(url=URL, data = formdata)
    soup = BeautifulSoup(markup=resp.text, features='html.parser')
    nodes1 = soup.select('h3 a')
    for j in nodes1:
        title.append(j.text)
    nodes2 = soup.select('p.summary')
    for k in nodes2:
        summary.append(k.text)
```

```python
pd.DataFrame(zip(title, summary),columns=['Title', 'Summary'])
```
