### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
header = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'}
resp = requests.get(url=URL, headers = header)
print(resp.text)
```
