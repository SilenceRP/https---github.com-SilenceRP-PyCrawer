## 1. 取出今天所有的新聞

```python
from selenium import webdriver
import requests
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path='c:/webdrivers/chromedriver.exe')
driver.get('https://www.ettoday.net/news/news-list.htm')
```

```python
content = []
dtime = []

nodes = driver.find_elements_by_css_selector('span.date')
for i in nodes:
    dtime.append(i.text)
    
nodes = driver.find_elements_by_css_selector('h3 a')
for i in nodes:
    content.append(i.text)
```

## 2. 取出現在時間兩小時內的新聞

```python
df = pd.DataFrame(zip(dtime, content), columns=['time', 'content'])
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from bs4 import BeautifulSoup
```

```python
formdata = {'offset':'11','tPage':'3','tFile':'20210530.xml',
            'tOt':'0','tSi':'100','tAr':'0'}
URL = 'https://www.ettoday.net/show_roll.php'            
```

```python
content = []
tdate = []

for i in ['27', '28', '29']:
    formdata['tFile'] = '202105{}.xml'.format(i)
    resp = requests.post(url=URL, data=formdata)
    soup = BeautifulSoup(markup=resp.text, features='html.parser')
    nodes1 = soup.select(selector = 'h3 a')
    nodes2 = soup.select(selector = 'span.date')
    for k in nodes1:
        content.append(k.text)
    for j in nodes2:
        tdate.append(j.text)
```

```python
pd.DataFrame(zip(tdate, content), columns=['Time', 'Content'])
```
