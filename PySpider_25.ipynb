```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.ettoday.net/news/20201004/1824032.htm'
resp = requests.get(url=URL)
```

```python
soup = BeautifulSoup(markup=resp.text ,features='html.parser')
nodes = soup.select(selector = '.subject_article')
for i in nodes:
    print(i.text)
```
